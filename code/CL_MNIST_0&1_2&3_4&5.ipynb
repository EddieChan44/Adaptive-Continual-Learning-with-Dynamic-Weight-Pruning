```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import Callback
import matplotlib.pyplot as plt
import pandas as pd

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("GPU is available and memory growth is set.")
    except RuntimeError as e:
        print(e)

# Enable mixed precision to speed up training
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Load the MNIST dataset
(img_train, label_train), (img_test, label_test) = tf.keras.datasets.mnist.load_data()

# Filter out images of digits "4" and "5" for training and testing datasets
train_filter_4_5 = (label_train == 4) | (label_train == 5)
test_filter_4_5 = (label_test == 4) | (label_test == 5)

img_train_4_5, label_train_4_5 = img_train[train_filter_4_5], label_train[train_filter_4_5]
img_test_4_5, label_test_4_5 = img_test[test_filter_4_5], label_test[test_filter_4_5]

# Convert labels to one-hot encoded vectors with 2 classes
label_train_4_5 = tf.keras.utils.to_categorical(label_train_4_5 - 4, num_classes=2)
label_test_4_5 = tf.keras.utils.to_categorical(label_test_4_5 - 4, num_classes=2)

# Filter out images of digits "0" and "1" for validation datasets
val_filter_0_1 = (label_test == 0) | (label_test == 1)
img_val_0_1, label_val_0_1 = img_test[val_filter_0_1], label_test[val_filter_0_1]
label_val_0_1 = tf.keras.utils.to_categorical(label_val_0_1, num_classes=2)

# Filter out images of digits "2" and "3" for additional validation dataset
val_filter_2_3 = (label_test == 2) | (label_test == 3)
img_val_2_3, label_val_2_3 = img_test[val_filter_2_3], label_test[val_filter_2_3]
label_val_2_3 = tf.keras.utils.to_categorical(label_val_2_3 - 2, num_classes=2)

# Function to convert grayscale images to RGB and resize them to 224x224
def preprocess_image(image, label):
    image = tf.expand_dims(image, axis=-1)  # Expand to add the channel dimension (28, 28) -> (28, 28, 1)
    image = tf.image.grayscale_to_rgb(image)  # Convert to RGB by repeating the channel (28, 28, 1) -> (28, 28, 3)
    image = tf.image.resize(image, (224, 224))  # Resize to 224x224
    image = image / 255.0  # Normalize to [0, 1]
    return image, label

# Convert the training and testing datasets into TensorFlow datasets
train_dataset_4_5 = tf.data.Dataset.from_tensor_slices((img_train_4_5, label_train_4_5))
test_dataset_4_5 = tf.data.Dataset.from_tensor_slices((img_test_4_5, label_test_4_5))

# Convert the validation datasets into TensorFlow datasets
val_dataset_0_1 = tf.data.Dataset.from_tensor_slices((img_val_0_1, label_val_0_1))
val_dataset_2_3 = tf.data.Dataset.from_tensor_slices((img_val_2_3, label_val_2_3))

# Apply the preprocessing function
train_dataset_4_5 = train_dataset_4_5.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
test_dataset_4_5 = test_dataset_4_5.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
val_dataset_0_1 = val_dataset_0_1.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
val_dataset_2_3 = val_dataset_2_3.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)

# Batch the datasets
batch_size = 32
train_dataset_4_5 = train_dataset_4_5.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
test_dataset_4_5 = test_dataset_4_5.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
val_dataset_0_1 = val_dataset_0_1.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
val_dataset_2_3 = val_dataset_2_3.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

# Load the pre-trained model
pretrained_model_path = 'CL_MNIST_vgg16_0&1_2&3.h5'
pretrained_model = load_model(pretrained_model_path)

# Display the pre-trained model summary
pretrained_model.summary()

# Compile the model with a new learning rate
fixed_learning_rate = 0.00001
pretrained_model.compile(optimizer=Adam(learning_rate=fixed_learning_rate),
                         loss='categorical_crossentropy',
                         metrics=['accuracy'])

# Custom callback to validate on "0" and "1" and "2" and "3" validation sets and save the model at the 10th epoch
class ZeroOneAndTwoThreeValidationCallback(Callback):
    def __init__(self, val_dataset_0_1, val_dataset_2_3):
        super(ZeroOneAndTwoThreeValidationCallback, self).__init__()
        self.val_dataset_0_1 = val_dataset_0_1
        self.val_dataset_2_3 = val_dataset_2_3
        self.val_loss_0_1 = []
        self.val_accuracy_0_1 = []
        self.val_loss_2_3 = []
        self.val_accuracy_2_3 = []

    def on_epoch_end(self, epoch, logs=None):
        val_loss_0_1, val_accuracy_0_1 = self.model.evaluate(self.val_dataset_0_1, verbose=0)
        self.val_loss_0_1.append(val_loss_0_1)
        self.val_accuracy_0_1.append(val_accuracy_0_1)
        print(f'\nEpoch {epoch + 1}: val_loss_0_1 = {val_loss_0_1}, val_accuracy_0_1 = {val_accuracy_0_1}\n')

        val_loss_2_3, val_accuracy_2_3 = self.model.evaluate(self.val_dataset_2_3, verbose=0)
        self.val_loss_2_3.append(val_loss_2_3)
        self.val_accuracy_2_3.append(val_accuracy_2_3)
        print(f'\nEpoch {epoch + 1}: val_loss_2_3 = {val_loss_2_3}, val_accuracy_2_3 = {val_accuracy_2_3}\n')

        # Save the model at the 10th epoch
        if epoch + 1 == 10:
            self.model.save('CL_MNIST_vgg16_0&1_2&3_4&5.h5')

# Initialize the custom callback
validation_callback = ZeroOneAndTwoThreeValidationCallback(val_dataset_0_1, val_dataset_2_3)

# Train the model without early stopping for 10 epochs
history = pretrained_model.fit(
    train_dataset_4_5,
    epochs=10,
    validation_data=test_dataset_4_5,
    callbacks=[validation_callback]
)

# Evaluate the model on the "4" and "5" test dataset
loss_4_5, accuracy_4_5 = pretrained_model.evaluate(test_dataset_4_5)
print(f"Test Loss on 4&5: {loss_4_5}")
print(f"Test Accuracy on 4&5: {accuracy_4_5}")

# Evaluate the model on the "0" and "1" validation dataset
loss_0_1, accuracy_0_1 = pretrained_model.evaluate(val_dataset_0_1)
print(f"Validation Loss on 0&1: {loss_0_1}")
print(f"Validation Accuracy on 0&1: {accuracy_0_1}")

# Evaluate the model on the "2" and "3" validation dataset
loss_2_3, accuracy_2_3 = pretrained_model.evaluate(val_dataset_2_3)
print(f"Validation Loss on 2&3: {loss_2_3}")
print(f"Validation Accuracy on 2&3: {accuracy_2_3}")

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 8))

epochs = range(1, len(history.history['accuracy']) + 1)

plt.subplot(2, 2, 1)
plt.plot(epochs, history.history['accuracy'], label='Train Accuracy')
plt.plot(epochs, history.history['val_accuracy'], label='Validation Accuracy on 4&5')
plt.title('Model Accuracy on 4&5')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, history.history['loss'], label='Train Loss')
plt.plot(epochs, history.history['val_loss'], label='Validation Loss on 4&5')
plt.title('Model Loss on 4&5')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, validation_callback.val_accuracy_0_1, label='Validation Accuracy on 0&1')
plt.title('Model Accuracy on 0&1')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, validation_callback.val_loss_0_1, label='Validation Loss on 0&1')
plt.title('Model Loss on 0&1')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(epochs, validation_callback.val_accuracy_2_3, label='Validation Accuracy on 2&3')
plt.title('Model Accuracy on 2&3')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, validation_callback.val_loss_2_3, label='Validation Loss on 2&3')
plt.title('Model Loss on 2&3')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Create a dataframe to list training and validation accuracy and loss for all epochs
history_df = pd.DataFrame(history.history)
history_df['epoch'] = history_df.index + 1
history_df['val_loss_0_1'] = validation_callback.val_loss_0_1
history_df['val_accuracy_0_1'] = validation_callback.val_accuracy_0_1
history_df['val_loss_2_3'] = validation_callback.val_loss_2_3
history_df['val_accuracy_2_3'] = validation_callback.val_accuracy_2_3
history_df = history_df[['epoch', 'accuracy', 'val_accuracy', 'loss', 'val_loss', 'val_loss_0_1', 'val_accuracy_0_1', 'val_loss_2_3', 'val_accuracy_2_3']]
print(history_df)
```
