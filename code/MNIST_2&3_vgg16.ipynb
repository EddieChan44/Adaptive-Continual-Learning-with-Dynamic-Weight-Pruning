```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import pandas as pd

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Set memory growth to avoid taking all the GPU memory
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("GPU is available and memory growth is set.")
    except RuntimeError as e:
        print(e)

# Enable mixed precision to speed up training
tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Load the MNIST dataset
(img_train, label_train), (img_test, label_test) = tf.keras.datasets.mnist.load_data()

# Filter out images of digits "2" and "3" for training and testing datasets
train_filter = (label_train == 2) | (label_train == 3)
test_filter = (label_test == 2) | (label_test == 3)

img_train, label_train = img_train[train_filter], label_train[train_filter]
img_test, label_test = img_test[test_filter], label_test[test_filter]

# Convert labels to one-hot encoded vectors with 2 classes
label_train = tf.keras.utils.to_categorical(label_train - 2, num_classes=2)
label_test = tf.keras.utils.to_categorical(label_test - 2, num_classes=2)

# Function to convert grayscale images to RGB and resize them to 224x224
def preprocess_image(image, label):
    image = tf.expand_dims(image, axis=-1)  # Expand to add the channel dimension (28, 28) -> (28, 28, 1)
    image = tf.image.grayscale_to_rgb(image)  # Convert to RGB by repeating the channel (28, 28, 1) -> (28, 28, 3)
    image = tf.image.resize(image, (224, 224))  # Resize to 224x224
    image = image / 255.0  # Normalize to [0, 1]
    return image, label

# Convert the training and testing datasets into TensorFlow datasets
train_dataset = tf.data.Dataset.from_tensor_slices((img_train, label_train))
test_dataset = tf.data.Dataset.from_tensor_slices((img_test, label_test))

# Apply the preprocessing function
train_dataset = train_dataset.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
test_dataset = test_dataset.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)

# Batch the datasets
batch_size = 32
train_dataset = train_dataset.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
test_dataset = test_dataset.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

# Define the VGG16-like architecture adjusted for MNIST with 224x224 input and kernel regularization
def build_vgg16_model(input_shape):
    model = Sequential()
    
    # Block 1
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape))
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Block 2
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Block 3
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Block 4
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Block 5
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(Conv2D(512, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Fully connected layers
    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(2, activation='softmax'))
    
    return model

# Build the model
model = build_vgg16_model((224, 224, 3))

# Display the model summary
model.summary()

# Compile the model
fixed_learning_rate = 0.00001
model.compile(optimizer=Adam(learning_rate=fixed_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model without early stopping for 10 epochs
history = model.fit(
    train_dataset,
    epochs=10,
    validation_data=test_dataset
)

# Save the trained model
model.save('MNIST_vgg16_2&3.h5')

# Evaluate the model
loss, accuracy = model.evaluate(test_dataset)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 4))

epochs = range(1, len(history.history['accuracy']) + 1)

plt.subplot(1, 2, 1)
plt.plot(epochs, history.history['accuracy'], label='Train Accuracy')
plt.plot(epochs, history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, history.history['loss'], label='Train Loss')
plt.plot(epochs, history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Create a dataframe to list training and validation accuracy and loss for all epochs
history_df = pd.DataFrame(history.history)
history_df['epoch'] = history_df.index + 1
history_df = history_df[['epoch', 'accuracy', 'val_accuracy', 'loss', 'val_loss']]
print(history_df)
```
