```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import Callback
import matplotlib.pyplot as plt
import pandas as pd

print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("GPU is available and memory growth is set.")
    except RuntimeError as e:
        print(e)

tf.keras.mixed_precision.set_global_policy('mixed_float16')

# Load the MNIST dataset
(img_train, label_train), (img_test, label_test) = tf.keras.datasets.mnist.load_data()

# Filter out images of digits "2" and "3" for training and testing datasets
train_filter_2_3 = (label_train == 2) | (label_train == 3)
test_filter_2_3 = (label_test == 2) | (label_test == 3)

img_train_2_3, label_train_2_3 = img_train[train_filter_2_3], label_train[train_filter_2_3]
img_test_2_3, label_test_2_3 = img_test[test_filter_2_3], label_test[test_filter_2_3]

# Convert labels to one-hot encoded vectors with 2 classes
label_train_2_3 = tf.keras.utils.to_categorical(label_train_2_3 - 2, num_classes=2)
label_test_2_3 = tf.keras.utils.to_categorical(label_test_2_3 - 2, num_classes=2)

# Filter out images of digits "0" and "1" for validation datasets
val_filter_0_1 = (label_test == 0) | (label_test == 1)

img_val_0_1, label_val_0_1 = img_test[val_filter_0_1], label_test[val_filter_0_1]

# Convert labels to one-hot encoded vectors with 2 classes for validation
label_val_0_1 = tf.keras.utils.to_categorical(label_val_0_1, num_classes=2)

# Function to convert grayscale images to RGB and resize them to 224x224
def preprocess_image(image, label):
    image = tf.expand_dims(image, axis=-1)  # Expand to add the channel dimension (28, 28) -> (28, 28, 1)
    image = tf.image.grayscale_to_rgb(image)  # Convert to RGB by repeating the channel (28, 28, 1) -> (28, 28, 3)
    image = tf.image.resize(image, (224, 224))  # Resize to 224x224
    image = image / 255.0  # Normalize to [0, 1]
    return image, label

# Convert the training and testing datasets into TensorFlow datasets
train_dataset_2_3 = tf.data.Dataset.from_tensor_slices((img_train_2_3, label_train_2_3))
test_dataset_2_3 = tf.data.Dataset.from_tensor_slices((img_test_2_3, label_test_2_3))

# Convert the validation dataset into TensorFlow dataset
val_dataset_0_1 = tf.data.Dataset.from_tensor_slices((img_val_0_1, label_val_0_1))

# Apply the preprocessing function
train_dataset_2_3 = train_dataset_2_3.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
test_dataset_2_3 = test_dataset_2_3.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)
val_dataset_0_1 = val_dataset_0_1.map(preprocess_image, num_parallel_calls=tf.data.experimental.AUTOTUNE)

# Batch the datasets
batch_size = 32
train_dataset_2_3 = train_dataset_2_3.shuffle(buffer_size=10000).batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
test_dataset_2_3 = test_dataset_2_3.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)
val_dataset_0_1 = val_dataset_0_1.batch(batch_size).prefetch(tf.data.experimental.AUTOTUNE)

# Load the pre-trained model
pretrained_model_path = 'MNIST_vgg16_0&1.h5'
pretrained_model = load_model(pretrained_model_path)

# Load the model with freezing rates
freezing_model_path = 'freezing_value_mag_0&1_0%_^1.h5'
freezing_model = load_model(freezing_model_path)

# Extract the freezing rates from the model weights
freezing_weights = [layer.get_weights() for layer in freezing_model.layers if layer.get_weights()]

# Define the fixed learning rate
fixed_learning_rate = 0.00001

# Define a custom optimizer with learning rate adjustment
class CustomAdam(Adam):
    def __init__(self, learning_rate, freezing_weights, *args, **kwargs):
        super(CustomAdam, self).__init__(learning_rate, *args, **kwargs)
        self.freezing_weights = freezing_weights
        self.learning_rate = learning_rate

    def _resource_apply_dense(self, grad, var, apply_state=None):
        var_key = self._var_key(var)
        freezing_weight = None

        for layer_weight in self.freezing_weights:
            if var.name in layer_weight[0].name:
                freezing_weight = layer_weight[0]
                break

        if freezing_weight is not None:
            learning_rate_new = self.learning_rate * freezing_weight
            return super(CustomAdam, self)._resource_apply_dense(grad, var, apply_state={
                "lr_t": learning_rate_new
            })
        else:
            return super(CustomAdam, self)._resource_apply_dense(grad, var, apply_state)

# Initialize the custom optimizer
custom_optimizer = CustomAdam(learning_rate=fixed_learning_rate, freezing_weights=freezing_weights)

# Display the pre-trained model summary
pretrained_model.summary()

# Compile the model with the custom optimizer
pretrained_model.compile(optimizer=custom_optimizer,
                         loss='categorical_crossentropy',
                         metrics=['accuracy'])

# Custom callback to validate on "0" and "1" validation set and save the model at the 10th epoch
class ZeroOneValidationAndSaveCallback(Callback):
    def __init__(self, validation_data):
        super(ZeroOneValidationAndSaveCallback, self).__init__()
        self.validation_data = validation_data
        self.val_loss_0_1 = []
        self.val_accuracy_0_1 = []

    def on_epoch_end(self, epoch, logs=None):
        val_loss, val_accuracy = self.model.evaluate(self.validation_data, verbose=0)
        self.val_loss_0_1.append(val_loss)
        self.val_accuracy_0_1.append(val_accuracy)
        print(f'\nEpoch {epoch + 1}: val_loss_0_1 = {val_loss}, val_accuracy_0_1 = {val_accuracy}\n')
        # Save the model at the 10th epoch
        if epoch + 1 == 10:
            self.model.save('MAG_MNIST_vgg16_0&1_2&3.h5')

# Initialize the custom callback
zero_one_validation_callback = ZeroOneValidationAndSaveCallback(val_dataset_0_1)

# Train the model without early stopping for 20 epochs
history = pretrained_model.fit(
    train_dataset_2_3,
    epochs=10,
    validation_data=test_dataset_2_3,
    callbacks=[zero_one_validation_callback]
)

# Evaluate the model on the "2" and "3" test dataset
loss_2_3, accuracy_2_3 = pretrained_model.evaluate(test_dataset_2_3)
print(f"Test Loss on 2&3: {loss_2_3}")
print(f"Test Accuracy on 2&3: {accuracy_2_3}")

# Evaluate the model on the "0" and "1" validation dataset
loss_0_1, accuracy_0_1 = pretrained_model.evaluate(val_dataset_0_1)
print(f"Validation Loss on 0&1: {loss_0_1}")
print(f"Validation Accuracy on 0&1: {accuracy_0_1}")

# Plot training & validation accuracy and loss
plt.figure(figsize=(12, 8))

epochs = range(1, len(history.history['accuracy']) + 1)

plt.subplot(2, 2, 1)
plt.plot(epochs, history.history['accuracy'], label='Train Accuracy')
plt.plot(epochs, history.history['val_accuracy'], label='Validation Accuracy on 2&3')
plt.title('Model Accuracy on 2&3')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(epochs, history.history['loss'], label='Train Loss')
plt.plot(epochs, history.history['val_loss'], label='Validation Loss on 2&3')
plt.title('Model Loss on 2&3')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(epochs, zero_one_validation_callback.val_accuracy_0_1, label='Validation Accuracy on 0&1')
plt.title('Model Accuracy on 0&1')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(epochs, zero_one_validation_callback.val_loss_0_1, label='Validation Loss on 0&1')
plt.title('Model Loss on 0&1')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# Create a dataframe to list training and validation accuracy and loss for all epochs
history_df = pd.DataFrame(history.history)
history_df['epoch'] = history_df.index + 1
history_df['val_loss_0_1'] = zero_one_validation_callback.val_loss_0_1
history_df['val_accuracy_0_1'] = zero_one_validation_callback.val_accuracy_0_1
history_df = history_df[['epoch', 'accuracy', 'val_accuracy', 'loss', 'val_loss', 'val_loss_0_1', 'val_accuracy_0_1']]
print(history_df)
```
